## 🔄 NCT CHART COMPARISON

So sánh biểu đồ NCT Top 50 giữa 2 ngày gần nhất

**Chức năng**:
  - Top 10 movements (↑↓)
  - New entries (bài hát mới vào chart)
  - Dropped songs (bài hát rơi khỏi chart)
  - Biggest gainers (bài hát tăng hạng nhiều nhất)
- **Cách dùng**: Chạy cell ở section 15 phía trên

```python
import pandas as pd
import os

INPUT_FILE = 'data/is_hit.csv'
OUTPUT_FILE = 'data/is_hit_list.csv'
    
print("Đã khai báo thư viện và tên file.")
print(f"Đang tải dữ liệu từ '{INPUT_FILE}'...")

if not os.path.exists(INPUT_FILE):
    print(f"LỖI: Không tìm thấy file '{INPUT_FILE}'.")
    print("Vui lòng chạy script '2_define_hit_labels.py' trước.")
else:
    df = pd.read_csv(INPUT_FILE)
    
    # --- LỌC BÀI "HIT" ---
    # Đây là logic chính: chỉ lấy các bài có nhãn = 1
    df_hits = df[df['label'] == 1].copy()
    df_hits_sorted = df_hits.sort_values(by='Total_Score', ascending=False) 
    print(f"Phân tích hoàn tất: Tìm thấy {len(df_hits)} bài 'True Hit' (Label=1).")
```

```python
if 'df_hits' in locals() and not df_hits.empty:
    print("Đang sắp xếp các bài 'Hit' theo điểm số (cao nhất xuống thấp nhất)...")
    
    # Sắp xếp theo Tổng Điểm
    df_hits_sorted = df_hits.sort_values(by='Total_Score', ascending=False)
    
    # (Tùy chọn) Chọn các cột quan trọng nhất để xem
    cols_to_show = [
        'track_name', 
        'artists', 
        'featured_artists',
        'Total_Score',
        'best_peak_rank',
        'total_platforms',
        'total_appearances',
        'spotify_release_date'
    ]
    
    # Lấy các cột tồn tại trong file của bạn
    final_cols_to_show = [col for col in cols_to_show if col in df_hits_sorted.columns]

    print("\n--- DANH SÁCH 'TRUE HIT' ---")
    # In ra toàn bộ danh sách "Hit"
    print(df_hits_sorted[final_cols_to_show].to_markdown(index=False, numalign="left", stralign="left"))
else:
    print("Không tìm thấy bài 'Hit' nào để hiển thị.")
```
